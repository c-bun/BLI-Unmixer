```python
def get_grid_images(plate_img, numColumn, numRow, x=35, y=79, height=15, width=15, spacing=3.5):
    """
    Takes raw IVIS luminescence images and divides into individual images based on a 96 well plate.
    """
    images = []
    for i in range(numRow):
        for j in range(numColumn):
            xOffset = int(j*(width+spacing))
            yOffset = int(i*(height+spacing))
            images.append(plate_img[y+yOffset:y+yOffset+height,x+xOffset:x+xOffset+width])
    return images
```

```python
im = get_grid_images(bromo,9,8)[-1]
im_r = im.shape[0]
im_c = im.shape[1]
zoom = im[4:im_r-4,4:im_c-4]
print(np.mean(im), np.mean(zoom))
plt.imshow(zoom)
plt.colorbar()
```

```python
def sq_crop(image, pixel_crop):
    im_r = image.shape[0]
    im_c = image.shape[1]
    return image[pixel_crop:im_r-pixel_crop,pixel_crop:im_c-pixel_crop]
```

```python
def norm(arr):
    n_arr = []
    if len(arr.shape) == 3:
        for i, sl in enumerate(arr):
            n_arr.append((sl-sl.min())/(sl.max()-sl.min()))
    else:
        n_arr = (arr-arr.min())/(arr.max()-arr.min())
    return np.array(n_arr)
```

```python
def lin_unmix(images, shape, e1, e2):
    """
    This will probably only work with two channels and two enzymes.
    """
    images = np.array([median_filter(im,5) for im in images])
    images_bkgd = [(im, np.mean(im[0:15,0:15])) for im in images]
    images = np.array([norm(im-bkgd) for im, bkgd in images_bkgd])
    ch1_grid_means = [np.mean(sq_crop(a,4)) for a in get_grid_images(images[0],*shape)]
    ch2_grid_means = [np.mean(sq_crop(a,4)) for a in get_grid_images(images[1],*shape)]

    X = np.array([(ch1_grid_means[e1],ch2_grid_means[e1]),
                 (ch1_grid_means[e2],ch2_grid_means[e2])])
    Y = np.identity(2)
    
    print(X,Y)
    
    reg = LinearRegression()
    reg.fit(X,Y)
    
    print(reg.coef_,reg.intercept_)
    
    unmix = reg.predict(images.reshape(256*256,2))
    return unmix.reshape(2,256,256)
```

Looks like the Gammon implementation is a bit different. They:

- divide all substrates by the max matrix
- produce a K matrix and return conditional value (rough estimate of unmixing quality)
- calculate the pseudoinverse of K
- apply to each z-stacked pixel

```python
stacked_raw = np.array([bromo, dluc])
stacked_norm = np.array([norm(x) for x in stacked_raw])
stacked_max = np.max(stacked_norm,axis=0)
stacked_global_norm = stacked_norm/stacked_max
```

```python
fig, ax = plt.subplots(nrows=4,ncols=2)

fig.set_size_inches(8,10)

ax[0,0].imshow(bromo)
ax[0,1].imshow(dluc)

ax[1,0].imshow(stacked_norm[0])
ax[1,1].imshow(stacked_norm[1])

ax[2,0].imshow(stacked_max)

ax[3,0].imshow(stacked_global_norm[0])
ax[3,1].imshow(stacked_global_norm[1])
```

```python
def find_K(images, layout_shape, e1, e2):
    """
    This will probably only work with two channels and two enzymes.
    """
    images = np.array([median_filter(im,4) for im in images])
    
    ch1_grid_means = [np.mean(sq_crop(a,4)) for a in get_grid_images(images[0],*layout_shape)]
    ch2_grid_means = [np.mean(sq_crop(a,4)) for a in get_grid_images(images[1],*layout_shape)]

    K = np.array([(ch1_grid_means[e1],ch1_grid_means[e2]),
                 (ch2_grid_means[e1],ch2_grid_means[e2])])
    return K
```

```python
K = find_K(stacked_global_norm, (9,8), 0, -1)
print(K)
```

```python
from numpy.linalg import pinv

K_inv = pinv(K)
print(K_inv)
```

```python
def plot_channels(stack):
    fig, axs = plt.subplots(nrows=1, ncols=stack.shape[0])
    fig.set_size_inches(10,6)
    for i, ax in enumerate(axs):
        ax.set_title("Channel {}".format(i+1))
        im = ax.imshow(stack[i])
        plt.colorbar(im, ax=ax)
```
