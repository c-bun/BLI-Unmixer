```python
def get_grid_images(plate_img, numColumn, numRow, x=35, y=79, height=15, width=15, spacing=3.5):
    """
    Takes raw IVIS luminescence images and divides into individual images based on a 96 well plate.
    """
    images = []
    for i in range(numRow):
        for j in range(numColumn):
            xOffset = int(j*(width+spacing))
            yOffset = int(i*(height+spacing))
            images.append(plate_img[y+yOffset:y+yOffset+height,x+xOffset:x+xOffset+width])
    return images
```

```python
def unmix_image(ch1, ch2, linearModel, scaler=None):
    x,y = ch1.shape
    flat1 = ch1.flatten()
    flat2 = ch2.flatten()
    flat = np.array([flat1,flat2]).T
    if scaler is not None:
        flat = scaler.fit_transform(flat)
    unmixed = linearModel.predict(flat)
    return np.reshape(unmixed,(x,y,2))
```
